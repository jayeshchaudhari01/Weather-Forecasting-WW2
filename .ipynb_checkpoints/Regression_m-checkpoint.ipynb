```python
import pandas as pd
import numpy as py
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('C:/Users/Jayesh/Project2/Temperature.csv')

df.describe()
```

## Data Cleaning

```python
df.columns
```

#### Removing/Dropping Invalid columns

```python
dropees=['PoorWeather', 'YR', 'MO', 'DA', 'PRCP', 'Unnamed: 11', 'Unnamed: 12',
       'Unnamed: 13', 'Unnamed: 14', 'Unnamed: 15', 'Unnamed: 16',
       'Unnamed: 17', 'Unnamed: 18', 'Unnamed: 19', 'Unnamed: 20',
       'Unnamed: 21', 'Unnamed: 22', 'Unnamed: 23']
df.drop(dropees, inplace= True, axis=1)
```

```python
df.head()
```

#### Replacing column names as more accurate format

```python
df.rename(columns={'STA': 'weather_station', 'Date':'date','MaxTemp':'max_temp',
                   'MinTemp':'min_temp', 'MeanTemp':'mean_temp','Snowfall':'snowfall'}, inplace=True)
df.info()
```

#### Changing null and invalid values  to 0

```python
df['snowfall'] = df['snowfall'].fillna(0)
df.info()
```

```python
df['snowfall'].value_counts()
```

Found a string value: '#VALUE!', changed it to a numeric 0

```python
df.replace(to_replace ="#VALUE!", value ="0", inplace=True)
df['snowfall'] = df['snowfall'].astype(float)
```

##### Now, let's remove the outlying data-points that are visible in the graph.

```python
df.drop(df[(df['max_temp']>10) & (df['min_temp']<-15)].index, inplace=True)
df.drop(df[(df['max_temp']>-10) & (df['min_temp']<-25)].index, inplace=True)
df.drop(df[(df['max_temp']<-2)  & (df['min_temp']>-5)].index, inplace=True)
```

```python
plt.figure(figsize=(10, 8))
sns.scatterplot(data=df, x='min_temp', y='max_temp').set(title='Minimum vs Maximum Temperature', 
                                                         xlabel='Min. Temp(C)', ylabel='Max. Temp(C)');
```

## Linear Regression


```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
model = LinearRegression()
```

### Defining x and y

```python
#x = df.drop(df.columns[[0, 1, 2, 5]], axis=1)
x=df[(df.columns[[3, 4]])].values
print(x)
#x are independent variables
```

```python
y=df['max_temp'].values
print(y)
#y is dependent variable
```

### Training the model 

```python
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size = 0.1, random_state=0)
```

```python
model.fit(train_x, train_y)
```

Checking the predict function for first row values

```python
model.predict([[22.222222, 23.888889]])
```

The predicted value (25.68107016) is nearly equal to the actual value (25.555556) of maximum temperature

```python
predicted_y=model.predict(test_x)
```

#### Relation between actual and predicted values

```python
from sklearn.metrics import r2_score
r2_score(test_y, predicted_y)
```

The r2_score is great and thus our prediction is significantly accurate.

#### Visualization #1

```python
plt.figure(figsize=(10, 10))
plt.scatter(test_y, predicted_y)
plt.xlabel('Actual max. temperature')
plt.ylabel('Predicted max. temperature')
plt.title('Actual v/s Predicted')
```

Almost all the actual and predicted values are pretty close except some.
